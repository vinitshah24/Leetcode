# Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of the two sorted arrays.

# The overall run time complexity should be O(log (m+n)).

 
## Example 1:
```python
Input: nums1 = [1,3], nums2 = [2]
Output: 2.00000
Explanation: merged array = [1,2,3] and median is 2.
```
## Example 2:
```python
Input: nums1 = [1,2], nums2 = [3,4]
Output: 2.50000
Explanation: merged array = [1,2,3,4] and median is (2 + 3) / 2 = 2.5.
```

## Constraints:
```python
nums1.length == m
nums2.length == n
0 <= m <= 1000
0 <= n <= 1000
1 <= m + n <= 2000
-106 <= nums1[i], nums2[i] <= 106
```

````python
def find_median(a1, a2):
    total_len = len(a1) + len(a2)
    merged_arr = []
    i = 0
    j = 0
    while i < len(a1) and j < len(a2):
        if a1[i] < a2[j]:
            merged_arr.append(a1[i])
            i += 1
        else:
            merged_arr.append(a2[j])
            j += 1
    merged_arr += a1[i:]
    merged_arr += a2[j:]
    median = 0.0
    mid = len(merged_arr) // 2
    if len(merged_arr) % 2 == 1:
        median = merged_arr[mid]
    else:
        median = (merged_arr[mid] + merged_arr[mid - 1]) // 2.0
    return median
````

````python
def find_median(a, b):
    # Check if the array "b" has more elements than "a"
    # because if "a" has more elements then there would be nothing
    # to select from "b" array. Binary search is applied on "a" array
    if len(a) == 1 and len(b) == 1:
        return float(a[0] + b[0]) // 2.0
    if len(a) > len(b):
        a, b = b, a
    low = 0
    high = len(a)
    total_elements = len(a) + len(b)
    while low <= high:
        a_left = (low + high) // 2
        b_left = (total_elements + 1) // 2 - a_left
        alml = -999 if a_left == 0 else a[a_left - 1]
        blml = -999 if b_left == 0 else b[b_left - 1]
        al = 999 if a_left == len(a) else a[a_left]
        bl = 999 if b_left == len(b) else b[b_left]
        # Apply Segregation
        if alml <= bl and blml <= al:
            median = 0
            # Even Elements
            if total_elements % 2 == 0:
                lmax = max(alml, blml)
                rmin = min(al, bl)
                median = (lmax + rmin) / 2.0
            # Odd Elements
            else:
                lmax = max(alml, blml)
                median = lmax
            return median
        elif alml > bl:
            # More elements need to be picked in left part of b array
            high = a_left - 1
        else:
            # blml > al:
            # More elements need to be picked in left part of a array
            low = a_left + 1
    return 0
````
