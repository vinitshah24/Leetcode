## Given the head of a linked list, remove the nth node from the end of the list and return its head.

## Example 1:

```python
Input: head = [1,2,3,4,5], n = 2
Output: [1,2,3,5]
```
## Example 2:
```python
Input: head = [1], n = 1
Output: []
```
## Example 3:
```python
Input: head = [1,2], n = 1
Output: [1]
```

## Constraints:
```python
The number of nodes in the list is sz.
1 <= sz <= 30
0 <= Node.val <= 100
1 <= n <= sz
```





````python
class ListNode:
    """ Definition for singly-linked list. """
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
````

````python
def removeNthFromEnd(head, n):
    # Create a dummy node so the left ends up target - 1 so we can assign left = left.next.next
    dummy = ListNode(0, head)
    left = dummy
    right = head
    # Now assign right to be exactly at the node we want to delete
    while n > 0 and right:
        right = right.next
        n -= 1
    # Move until right reaches NULL
    while right:
        right = right.next
        left = left.next
    # Now delete the node
    left.next = left.next.next
    # Return the head back
    return dummy.next
````

````python
head = [1, 2, 3, 4, 5]
n = 2

a = ListNode(1)
b = ListNode(2)
c = ListNode(3)
d = ListNode(4)
e = ListNode(5)

a.next = b
b.next = c
c.next = d
d.next = e

print("INITIAL NODE:")
curr = a
while curr is not None:
    print(curr.val, end=" ")
    curr = curr.next

print("\n\nNODE N DELETION:")
new_head = removeNthFromEnd(a, n=2)
curr = new_head
while curr is not None:
    print(curr.val, end=" ")
    curr = curr.next
````
